```python
import statsmodels.formula.api as smf
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    "X": [1, 2, 3, 4, 5],
    "y": [2.1, 4.2, 6.1, 8.3, 10.4]
})
model1 = smf.ols('y ~ X', data=df)
result = model1.fit()
print(result.summary())

ypred = result.predict(df["X"])
print(ypred)

print(f"Coefficients:\n{result.params}")
print(f"P-values:\n{result.pvalues}")
print(f"R-squared:\n{result.rsquared}")

plt.figure(figsize=(10, 5))
plt.plot(df['X'], df['y'], 'o', label='Actual Data')
plt.plot(df['X'], ypred, 'r', linewidth=2, label='Regression Line')  # Regression line
plt.xlabel('X values')
plt.ylabel('y values')
plt.title('Linear Regression Model using `statsmodels`')
plt.legend()
plt.grid(True)
plt.show()
```
