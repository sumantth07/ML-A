```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x = np.linspace(-5, 10, 100)
pdf = norm.pdf(x, loc=2, scale=1.5)

plt.plot(x, pdf, label='Normal PDF')
plt.title('Probability Density Function (Normal)')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```
